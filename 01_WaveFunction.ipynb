# Wave Function

The state of a quantum particle at time $t$ &ndash; neglecting *spin* and other *internal* variables &ndash; is postulated to be fully defined by a complex valued function of space coordinates, the so-called *wave function*:

$$\Psi(\boldsymbol{x}, t)$$ 


## Schrödinger Equation

The time evolution of a quantum particle's state is governed by the Schrödinger partial differential equation:

$$i \hslash \frac{\partial \Psi}{\partial t} = - \frac{\hslash^{2}}{2m} \frac{\partial^{2} \Psi}{\partial x^{2}} + V \: \Psi$$

where $\hslash$ is Planck's constant $h$ divided by $2\pi$:

$$\hslash = \frac{h}{2\pi} = 1.054572 \times 10^{-34} Joule \cdot second. $$

### Statistical Interpretation

The squared module of the wave function &ndash; $\Psi^{*}\Psi = \|\Psi(\bar{\boldsymbol{x}}, t)\|^{2}$ &ndash; is interpreted as the *probability* that in a measurement of *position* the particle be found at $\boldsymbol{x} = \bar{\boldsymbol{x}}$. 

Experiments show that after a measurement of position, successive measurements performed immediately after produce an identical result. This is interpreted by saying that *position measurements collapse the wave function to a very peaked distribution of $\Psi$ around the measured value*. Before performing measurements not only the shape of $\Psi$ is not known but even assuming that the particle be at some unknown position is contradicted by experimental evidence (Bell inequalities).    

### Normalization
The **linearity** of the Schrödinger equation ensures that if it is satisfied by a wave function $\Psi$ then $k \Psi$ is an acceptable solution as well. Therefore, any solution is determined up to an arbitrary multiplicative (complex) constant.
However, if $\|\Psi(\bar{\boldsymbol{x}}, t)\|^{2}$ is to be interpreted as a *probability density*, then its integral over all space must be $1$. We thus require for a valid $\Psi$ in three-dimensional space (cartesian coordinates)

$$ \int_{- \infty}^{+ \infty} dx \int_{- \infty}^{+ \infty} dy \int_{- \infty}^{+ \infty} dz \: \|\Psi(x,y,z; t)\|^{2} \: = \: 1$$

Note that **normalization only determines the wave function modulus**. Multiplying a normalized wave function by a *phase* factor &ndash; an arbitrary complex number of unit modulus &ndash; is always possible and it has no consequences. 


### Probability is conserved
For the statistical interpretation to be logically consistent the time evolution of a normalized quantum state as dictated by the Schrödinger partial differential equation must not spoil the wave function normalization property. 

That is, we must prove that (in one spatial dimension, for the sake of space) 

$$ 0 = \frac{d}{dt} \, \int_{- \infty}^{+ \infty} \, \|\Psi(x, t)\|^{2} dx \, = \, \int_{- \infty}^{+ \infty} \, \frac{\partial}{\partial t}  \|\Psi(x, t)\|^{2} dx \, =   \, \int_{- \infty}^{+ \infty} \, \frac{\partial}{\partial t}  (\Psi^{*}\Psi) dx\, = \, \int_{- \infty}^{+ \infty} \, \left[ \Psi^{*} \frac{\partial \Psi}{\partial t} + \frac{\partial \Psi^{*}}{\partial t} \Psi \right] dx$$

By the Schrödinger equation

$$ \frac{\partial \Psi}{\partial t} = \frac{i \hslash}{2m} \frac{\partial^{2} \Psi}{\partial x^{2}} - \frac{i}{\hslash} V \, \Psi$$

By taking the complex conjugate of the former equation, assuming the potential $V$ to be a *real* valued function:

$$ \frac{\partial \Psi^{*}}{\partial t} = - \frac{i \hslash}{2m} \frac{\partial^{2} \Psi^{*}}{\partial x^{2}} + \frac{i}{\hslash} V \, \Psi^{*}$$

Now, summing side by side the first of the two preceding equations multiplied by $\Psi^{*}$ and the second one multiplied by $\Psi$ we obtain:

$$\frac{\partial \left| \Psi \right|^{2}} {\partial t} = \, \left[ \Psi^{*} \frac{\partial \Psi}{\partial t} + \frac{\partial \Psi^{*}}{\partial t} \Psi \right] = \, \frac{i \hslash}{2m} \left( \Psi^{*}\frac{\partial^{2} \Psi}{\partial x^{2}} - \frac{\partial^{2} \Psi^{*}}{\partial x^{2}} \Psi \right) = \, \frac{\partial}{dx} \left[  \frac{i \hslash}{2m} \left(  \Psi^{*} \frac{\partial \Psi}{\partial x} - \frac{\partial \Psi^{*}}{\partial x} \Psi \right) \right]$$

This result allows us to evaluate the integral at the right hand side of the equation opening this section:

$$ \int_{- \infty}^{+ \infty} \, \left[ \Psi^{*} \frac{\partial \Psi}{\partial t} + \frac{\partial \Psi^{*}}{\partial t} \Psi \right] dx = \int_{- \infty}^{+ \infty} \, \frac{\partial}{dx} \left[  \frac{i \hslash}{2m} \left(  \Psi^{*} \frac{\partial \Psi}{\partial x} - \frac{\partial \Psi^{*}}{\partial x} \Psi \right) \right] dx = \left[  \frac{i \hslash}{2m} \left(  \Psi^{*} \frac{\partial \Psi}{\partial x} - \frac{\partial \Psi^{*}}{\partial x} \Psi \right) \right]_{- \infty}^{+ \infty} $$

That this must vanish follows from $\Psi$ and its derivative vanishing at infinity quicker than $\sqrt{x}$, as required for the wave function to be normalized.

### Probability current

When generalizing the above derivation to three-dimensional space this result takes the following form: 

$$ \int_{V} \, \frac{\partial}{\partial t}  \|\Psi(\boldsymbol{r}, t)\|^{2} d \boldsymbol{r} =  \, \frac{i \hslash}{2m}  \int_{V} \, \left( \Psi^{*}\Delta \Psi  - \Psi \Delta \Psi^{*} \right) d \boldsymbol{r}  =  \, \frac{i \hslash}{2m} \int_{S} \left( \Psi^{*}\nabla \Psi  - \Psi \nabla \Psi^{*} \right) d \boldsymbol{s}$$


where $\Delta f$ is the **laplacian** of $f$ (a scalar) and $\nabla f$ is the **gradient** of $f$ (a vector).

By defining 

$$P(\boldsymbol{r}, t) \equiv \|\Psi(\boldsymbol{r}, t)\|^{2}$$ and 

$$\boldsymbol{J} \equiv - \frac{i \hslash}{2m} \left( \Psi^{*}\nabla \Psi  - \Psi \nabla \Psi^{*} \right)$$ 

the previous result can be cast in the form of an *equation of continuity* analogous to the one encountered in fluid dynamics: 


$$ \int_{V} \frac{\partial P(\boldsymbol{r}, t)}{\partial t} + \, \nabla \cdot \boldsymbol{J} \,  d \boldsymbol{r} = 0 $$

By transforming the *divergence* term in the volume integral into a surface integral, the equation takes the suggestive form 

$$ \int_{V} \frac{\partial P(\boldsymbol{r}, t)}{\partial t} + \, \int_{S} \boldsymbol{J} \cdot d \boldsymbol{s} = 0 $$ 

which can be interpreted by saying that *any increment/decrement of the probability enclosed within a control volume $V$ is associated with an equal amount flowing inside/outside the volume through its surface*.  

In summary, we have demonstrated is that if we take as the control volume the whole space, then the total probability associated to the wave function is conserved (there is no flowing of probability at the infinitely distant boundaries). 

The vector $\boldsymbol{J}$ can be interpreted as a **probability current**. Moreover, since the control volume $V$ can be arbitrarily chosen, we can use the continuity equation in its differential form:

$$ \frac{\partial P(\boldsymbol{r}, t)}{\partial t} + \, \nabla \cdot \boldsymbol{J} = 0$$
